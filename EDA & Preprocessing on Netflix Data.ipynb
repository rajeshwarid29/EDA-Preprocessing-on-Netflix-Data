# Count of movie and tv show based on type column

```python
data["type"].value_counts()
```

```python
genre_series = data['listed_in'].str.split(',').explode().str.strip()
print (genre_series)
```

```python
genre_counts = genre_series.value_counts()
genre_counts
```

# Visualizing the output of genre series

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
genre_counts.plot(kind = "bar",color = "red")
plt.title("Distribution of Genres")
plt.xlabel("Genres")
plt.ylabel("Number of Titles")
plt.show()
```

# Distribution of content based on country

```python
country_content = data["country"].str.split(',').explode().str.strip()
print (country_content)
```

```python
country_count = country_content.value_counts()
country_count
```

# To Identify Countries with the Highest and Lowest Production


```python
highest_production = country_count.idxmax()
lowest_production = country_count.idxmin()

print(f"Highest production country:{highest_production}")
print(f"Lowest production country:{lowest_production}")
```

# Top 10 movies and shows based on rating

```python
#converting rating column to numeric from categorical
data["rating"] = pd.to_numeric(data["rating"], errors = "coerce")
```

```python
#top 10 movies shows based on rating
top_10 = data.nlargest(10, "rating")
print(top_10[["title","type","rating"]])
```

# Top 10 directors on the basis of count of titles

```python
top10_directors = data["director"].value_counts().nlargest(10)
print(f"Top 10 directors based on number of titles:\n {top10_directors}")
```

```python
plt.figure(figsize=(10, 6))
top10_directors.plot(kind='bar', color='purple')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Directors')
plt.ylabel('Number of Titles')
plt.xticks(rotation=60, ha='right')
plt.show()
```

# Top 10 cast on the basis of count of titles

```python
top_10 = data["cast"].str.split(',').explode().str.strip()
top_10
```

```python
top10_cast = top_10.value_counts().nlargest(10)
print("Top 10 cast based on number of titles:\n",top10_cast)
```

```python
plt.figure(figsize=(12, 6))
top10_cast.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cast Members on Netflix')
plt.xlabel('Cast Member')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
